```python
import cv2  
import matplotlib.pyplot as plt  
img = cv2.imread('soya.jpeg')  
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  
plt.show()
```

```python
import cv2  
import matplotlib.pyplot as plt  
img = cv2.imread('soya.jpeg')  
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), cmap='gray', vmin=0, vmax=255) 
plt.show()
```
