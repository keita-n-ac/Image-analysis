1. OpenCVライブラリを使わずに，解像度3×3で以下の画像を描きなさい


```python
import numpy as np
import matplotlib.pyplot as plt # ライブラリの導⼊
image = np.full((3, 3, 3), 255, np.uint8) # 黒画像を作成
image[0][2][1] = 0 # 高さ0, 幅2の画素の緑要素を0にする
image[0][2][2] = 0 # 高さ0, 幅2の画素の青要素を0にする
image[1][1][0] = 0 # 高さ1, 幅1の画素の赤要素を0にする
image[1][1][2] = 0 # 高さ1, 幅1の画素の青要素を0にする
image[2][0][0] = 0 # 高さ2, 幅0の画素の赤要素を255にする
image[2][0][1] = 0 # 高さ2, 幅0の画素の緑要素を255にする
plt.imshow(image)
plt.show()
```

2. OpenCVライブラリを使って，解像度600×400で，日本国旗を描きなさい

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt # ライブラリの導⼊
image = np.full((400, 600, 3), 255, np.uint8) # 白画像を作成
cv2.circle(image, (300, 200), 100, (255, 0, 0), -1)
plt.imshow(image)
plt.show()
```

3. OpenCVライブラリを使って，解像度800×600で，ドイツ国旗を描きなさい

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt # ライブラリの導⼊
image = np.zeros((600, 800, 3), np.uint8) # 黒画像を作成
cv2.rectangle(image, (0, 200, 800, 200), (255, 0, 0), -1)
cv2.rectangle(image, (0, 400, 800, 200), (255, 255, 0), -1)
plt.imshow(image)
plt.show()
```
