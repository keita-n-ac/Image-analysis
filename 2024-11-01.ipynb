### 画素値のヒストグラム作成（基本: OpenCVの機能だけで表示）
- グレースケール画像の場合
    - opencvをimportする: ```import cv2```
    - matplotlibをimportする: ```import matplotlib.pyplot as plt```
    - 白黒変換した画像を変数に代入する
    - 白黒変換した画像に対して，```ヒストグラムデータ変数 = cv2.calcHist([白黒変換した画像変数], [0], None, [256], [0,256])```とする
    - ヒストグラムデータがヒストグラムデータ変数に代入される
    - ```plt.plot(ヒストグラムデータ変数)```と```plt.show()```を使用してヒストグラムを表示する

````python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
hist = cv2.calcHist([image], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
````

- ```print(hist[画素値])```でその画素値の度数がわかる

````python
print(hist[0])
print(hist[32])
print(hist[64])
print(hist[128])
print(hist[255])
````

- ```plt.plot(hist, color=色文字列)```とすることで，描画の色を変更できる
    - 'r' : 赤
    - 'g' : 緑
    - 'b' : 青
    - 'c' : シアン（水色）
    - 'm' : マゼンタ（赤紫）
    - 'y' : 黄
    - 'k' : 黒
    - 'w' : 白
- ```plt.plot(hist, label=ラベル文字列)```とすることで，グラフに名前（凡例）を付けることができる
    - **名前は英数字で書くこと（matplotlibは日本語を扱えない）**
    - 凡例を表示する場合，```plt.legend()```を付ける
- グラフのタイトル，ラベルを指定する
  - **英数字で書くこと（matplotlibは日本語を扱えない）**
  - ```plt.title('文字列')``` : グラフのタイトルを指定した文字列にする
  - ```plt.xlable('文字列')``` : x軸のラベルを指定した文字列にする
  - ```plt.ylable('文字列')``` : y軸のラベルを指定した文字列にする

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
hist = cv2.calcHist([image], [0], None, [256], [0,256])
plt.title('Histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.plot(hist, color='k', label='Frequency')
plt.legend()
plt.show()
````

### 画素値のヒストグラム作成（基本: OpenCVの機能だけで表示）
- カラー画像の場合
    - opencvをimportする: ```import cv2```
    - matplotlibをimportする: ```import matplotlib.pyplot as plt```
    - 画像データを変数に代入する
    - BGRからRGBに変換した画像に対して，```ヒストグラムデータ変数 = cv2.calcHist([白黒変換した画像変数], [i], None, [256], [0,256])```とする
        - iが0: **赤**の画素値
        - iが1: **緑**の画素値
        - iが2: **青**の画素値
    - ヒストグラムデータがヒストグラムデータ変数に代入される
    - ```plt.plot(ヒストグラムデータ変数)```と```plt.show()```を使用してヒストグラムを表示する
```

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# 赤の画素値に関するヒストグラム
hist_red = cv2.calcHist([image], [0], None, [256], [0,256])

# 緑の画素値に関するヒストグラム
hist_green = cv2.calcHist([image], [1], None, [256], [0,256])

# 青の画素値に関するヒストグラム
hist_blue = cv2.calcHist([image], [2], None, [256], [0,256])

# 3つのヒストグラムを色を変えて表示する
plt.plot(hist_red, color='r', label='Red Frequency')
plt.plot(hist_green, color='g', label='Green Frequency')
plt.plot(hist_blue, color='b', label='Blue Frequency')

plt.title('Histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')

plt.legend()
plt.show()
````

### 画素値のヒストグラム作成（応用: Numpyのライブラリの機能を使って表示）
- Numpyライブラリで画素値を扱うと，画素値に対して一括処理を行うことができる
- グレースケール画像の場合
    - opencvをimportする: ```import cv2```
    - numpyをimportする: ```import numpy as np```
    - matplotlibをimportする: ```import matplotlib.pyplot as plt```
    - 白黒変換した画像を変数に代入する
    - 白黒変換した画像に対して，```ヒストグラムデータ変数，階級変数 = np.histogram(白黒変換した画像.ravel(), bins=256, range=(0, 256))```とする
    - ヒストグラムデータがヒストグラムデータ変数に代入される
    - ```plt.plot(ヒストグラムデータ変数)```と```plt.show()```を使用してヒストグラムを表示する

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー

# ヒストグラムの取得
hist, bins = np.histogram(image.ravel(), bins=256, range=(0, 256))

# ヒストグラムの表示
plt.plot(hist)
plt.show()
````

### 画素値のヒストグラム作成（応用: Numpyのライブラリの機能を使って表示）
- Numpyライブラリで画素値を扱うと，画素値に対して一括処理を行うことができる
- カラー画像の場合
    - opencvをimportする: ```import cv2```
    - numpyをimportする: ```import numpy as np```
    - matplotlibをimportする: ```import matplotlib.pyplot as plt```
    - 画像データを変数に代入し，BGRからRGBに変換する
    - 以下の処理を行うことで，RGB各画素のデータを変数に代入できる
        - ```赤画素要素変数 = 画像データ変数[:, :, 0]```
        - ```緑画素要素変数 = 画像データ変数[:, :, 1]```
        - ```青画素要素変数 = 画像データ変数[:, :, 2]```
    - ヒストグラムデータ変数，階級変数 = np.histogram(画素要素変数.ravel(), bins=256, range=(0, 256))```とする
    - ヒストグラムデータがヒストグラムデータ変数に代入される
    - ```plt.plot(ヒストグラムデータ変数)```と```plt.show()```を使用してヒストグラムを表示する

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR → RGB

# RGB画素値の取り出し
red_data = image[:, :, 0]
green_data = image[:, :, 1]
blue_data = image[:, :, 2]

# ヒストグラムの取得
red_hist, red_bins = np.histogram(red_data.ravel(), bins=256, range=(0, 256))
green_hist, green_bins = np.histogram(green_data.ravel(), bins=256, range=(0, 256))
blue_hist, blue_bins = np.histogram(blue_data.ravel(), bins=256, range=(0, 256))

# ヒストグラムの表示
plt.plot(red_hist, color='r', label='Red Frequency')
plt.plot(green_hist, color='g', label='Green Frequency')
plt.plot(blue_hist, color='b', label='Blue Frequency')

plt.legend()
plt.title('Histogram')
plt.show()
````

### ヒストグラム均一化
- opencvをimportする: ```import cv2```
- matplotlibをimportする: ```import matplotlib.pyplot as plt```
- 白黒変換した画像を変数に代入する
- 白黒変換した画像に対して，```ヒストグラム均一化変数 = cv2.equalizeHist(白黒変換した画像変数)```とする
    - ```cv2.equalizeHist```がヒストグラム均一化する命令
    - ヒストグラム均一化変数にヒストグラム均一化された画素値が代入される
- ```plt.imshow(ヒストグラム均一化変数)```と```plt.show()```を使用してヒストグラムを表示する

````python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
plt.imshow(image)
plt.gray()
plt.show()
````

````python
# ヒストグラム均一化
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
after_image = cv2.equalizeHist(image) # ヒストグラム均一化
plt.imshow(after_image)
plt.gray()
plt.show()
````

````python
# ヒストグラム均一化した画像における画素値分布
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
hist = cv2.calcHist([image], [0], None, [256], [0,256])
plt.plot(hist)
plt.gray()
plt.show()
````

````python
# ヒストグラム均一化した画像における画素値分布
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
after_image = cv2.equalizeHist(image) # ヒストグラム均一化
hist = cv2.calcHist([after_image], [0], None, [256], [0,256])
plt.plot(hist)
plt.gray()
plt.show()
````
