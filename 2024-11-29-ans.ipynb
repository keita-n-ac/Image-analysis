### 演習1
- iTossにある```boston.jpeg```に対して，各2値化処理を行いなさい．また，2値化処理を行った画像の画素値ヒストグラムを作成しなさい．（1行に，2値化画像とヒストグラムを出力すること）
- 2値化処理は，以下の方法で行うこと（出力は縦3分割，横2分割になる）
    - ```cv2.THRESH_BINARY```  
    - ```cv2.THRESH_TRUNC``` 
    - ```cv2.THRESH_TOZERO```

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('boston.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 12)) # 横10インチ，縦12インチ

# 二値化を行う
threshold = 127 # しきい値
value, after_image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
plt.subplot(321) # 縦1分割，横2分割の1番目に描画
plt.title('cv2.THRESH_BINARY')
plt.imshow(after_image)
plt.gray()

# ヒストグラム作成
plt.subplot(322) # 縦3分割，横2分割の2番目に描画
hist = cv2.calcHist([after_image], [0], None, [256], [0,256])
plt.title('Histogram')
plt.plot(hist)

# 二値化を行う
threshold = 127 # しきい値
value, after_image = cv2.threshold(image, threshold, 255, cv2.THRESH_TRUNC)
plt.subplot(323) # 縦3分割，横2分割の3番目に描画
plt.title('cv2.THRESH_TRUNC')
plt.imshow(after_image)
plt.gray()

# ヒストグラム作成
plt.subplot(324) # 縦3分割，横2分割の4番目に描画
hist = cv2.calcHist([after_image], [0], None, [256], [0,256])
plt.title('Histogram')
plt.plot(hist)


# 二値化を行う
threshold = 127 # しきい値
value, after_image = cv2.threshold(image, threshold, 255, cv2.THRESH_TOZERO)
plt.subplot(325) # 縦3分割，横2分割の5番目に描画
plt.title('cv2.THRESH_TOZERO')
plt.imshow(after_image)
plt.gray()

# ヒストグラム作成
plt.subplot(326) # 縦3分割，横2分割の6番目に描画
hist = cv2.calcHist([after_image], [0], None, [256], [0,256])
plt.title('Histogram')
plt.plot(hist)

# 全体を表示
plt.show()
````

### 演習2
- iTossにある```boston.jpeg```に対して，マスク画像を作成し，マスク処理を行いなさい．黒背景に長方形の白抜き画像をマスク画像とすること．
- サンプルプログラムにあるように，3つの画像（入力画像である```boston.jpeg```，マスク画像，マスク処理後の出力画像）を1行に並べて出力すること

````python
# グレースケール画像の場合
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('boston.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 2つ表示する
plt.figure(figsize=(14, 12)) # 横14インチ，縦12インチ

# 1つ目表示
plt.subplot(131) # 縦1分割，横3分割の1番目に描画
plt.title('Input image')
plt.imshow(image)

# 2つ目表示
plt.subplot(132) # 縦1分割，横3分割の2番目に描画
height = image.shape[0] # 縦ピクセル数
width = image.shape[1] # 横ピクセル数
mask = np.full((height, width, 3), 0, np.uint8) # マスク画像の黒背景
cv2.rectangle(mask, (150, 70, 280, 150), (255, 255, 255), -1) # 白抜き
plt.imshow(mask)
plt.gray()

# 3つ目表示
plt.subplot(133) # 縦1分割，横3分割の3番目に描画
after_image = cv2.bitwise_and(image, mask) # マスク処理を行う
plt.title('Output image')
plt.imshow(after_image)

plt.show()
````
