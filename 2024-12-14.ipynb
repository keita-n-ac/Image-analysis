### ルックアップテーブルの作成
- 変換前は ```np.arange(256)```を使用する

````python
import numpy as np
import matplotlib.pyplot as plt
before_pixel_value = np.arange(256)
print(before_pixel_value)
plt.plot(before_pixel_value)
plt.show()
````

````python
# 逆順にする場合, before_pixel_value[::-1]を使う
after_pixel_value = before_pixel_value[::-1] #[::-1]が逆順という命令
print(after_pixel_value)
plt.plot(after_pixel_value)
plt.show()
````

### a, bを指定し，画素値0からa-1を0，aからbを255，b+1から255までを0にする（新規内容）
- 以下のプログラムを書く
    ```np.where((before_pixel_value >= a) & (before_pixel_value <= b), 255, 0)```
- ```np.where((ピクセル変数 >= a) & (b <= ピクセル変数), x, y)``` と書くことで，画素値a以上b以下をxにし，そうで無い場合yにすると言う命令になる

````python
# np.whereを使用する
a = 50
b = 205
after_pixel_value = np.where((before_pixel_value >= a) & (before_pixel_value <= b), 255, 0)
print(after_pixel_value)
plt.plot(after_pixel_value)
plt.show()
````

### 折れ線トーングラフ
- 入力画像の画素値0から$a$を0にし，入力画像の画素値$b$から255を255にする画像変換
- 入力画像の画素値$a$から$b$を直線（1次関数）的に変換する
0. $a$と$b$の値を設定する（$a < b$）
1. まず始めに，以下の式で，画素値を変更する（この操作で$a$から$b$が求まる）
```python
after_pixel_value = 255 * (before_pixel_value - a) / (b - a)
```
2. 次に，求めた変換後の画素値を0から255に収める（この操作で0から$a$，$b$から255が求まる）
```
after_pixel_value = np.clip(after_pixel_value, 0, 255)
```
- ```np.clip(画素値変数, 最小値，最大値)```とかくことで，最小値より小さい値を最小値にし，最大値より大きい値を最大値にする
- この操作の後に，以下の処理を行う（全ての計算が終わった後に行う）
```
after_pixel_value = after_pixel_value.astype(np.uint8)
```

````python
# 折れ線トーンカーブ
import numpy as np
import matplotlib.pyplot as plt

a = 55
b = 200
before_pixel_value = np.arange(256)
after_pixel_value = 255 * (before_pixel_value - a) / (b - a)
after_pixel_value = np.clip(after_pixel_value, 0, 255)
after_pixel_value = after_pixel_value.astype(np.uint8)

plt.plot(after_pixel_value)
plt.show()
````

### 折れ線トーンカーブの逆順
- ```[::-1]```を使用すればよい

````python
# 折れ線カーブの逆順
import matplotlib.pyplot as plt

a = 55
b = 200
before_pixel_value = np.arange(256)
after_pixel_value = 255 * (before_pixel_value - a) / (b - a)
after_pixel_value = np.clip(after_pixel_value, 0, 255)
after_pixel_value = after_pixel_value.astype(np.uint8)
after_pixel_value = after_pixel_value[::-1] # [::-1]をすることで逆順にする

plt.plot(after_pixel_value)
plt.show()
````

### 折れ線トーングラフ（逆順）
- 入力画像の画素値0から$a$を255にし，入力画像の画素値$b$から255を0にする画像変換
- 入力画像の画素値$a$から$b$を直線（1次関数）的に変換する
0. $a$と$b$の値を設定する（$a < b$）
1. 逆順にすることを考慮して，
- 入力画像の画素値0から$c$を0にし，入力画像の画素値$d$から255を255にする画像変換
- 入力画像の画素値$c$から$d$を直線（1次関数）的に変換する
- ただし，$c = 255 - b$, $d = 255 - a$
2. まず始めに，以下の式で，画素値を変更する（この操作で$c$から$d$が求まる）
```python
after_pixel_value = 255 * (before_pixel_value - c) / (d - c)
```
3. 次に，求めた変換後の画素値を0から255に収める（この操作で0から$c$，$d$から255が求まる）
```
after_pixel_value = np.clip(after_pixel_value, 0, 255)
```
- ```np.clip(画素値変数, 最小値，最大値)```とかくことで，最小値より小さい値を最小値にし，最大値より大きい値を最大値にする
- この操作の後に，以下の処理を行う（全ての計算が終わった後に行う）
```
after_pixel_value = after_pixel_value.astype(np.uint8)
```
4. 最後に，```[::-1]```を使用して逆順にする
```
after_pixel_value = after_pixel_value.astype[::-1]
```


````python
# 折れ線カーブの逆順
import numpy as np
import matplotlib.pyplot as plt

a = 25
b = 200
c = 255 - b
d = 255 - a

before_pixel_value = np.arange(256)
after_pixel_value = 255 * (before_pixel_value - c) / (d - c)
after_pixel_value = np.clip(after_pixel_value, 0, 255)
after_pixel_value = after_pixel_value.astype(np.uint8)
after_pixel_value = after_pixel_value[::-1] # [::-1]をすることで逆順にする

print(after_pixel_value)
plt.plot(after_pixel_value)
plt.show()
````

### 台形トーンカーブ
- 3つの直線の組み合わせで考える
 1. 画素値0から画素値aを0から255で直線変化
 2. 画素値aから画素値bを255にする
 3. 画素値bから画素値255を255から0に直線変化


#### Step:1 画素値0から画素値aを0から255で直線変化
 - 画素値0から画素値aに対して，```255/a * 画素値```で計算できる
 - 画素値0から画素値aは```画素値変数[0:a+1]```でアクセスできる
  - ```[i:j]```はiからjの手前（つまりj - 1）までという意味
  - ```[0:a+1]```は0からa+1の手前（つまりa）までという意味

````python
# 台形トーンカーブ
import numpy as np
import matplotlib.pyplot as plt

a = 25
before_pixel_value = np.arange(256)
after_pixel_value1 = 255 / a * before_pixel_value
after_pixel_value1 = np.clip(after_pixel_value1, 0, 255)
after_pixel_value1 = after_pixel_value1.astype(np.uint8)

plt.plot(after_pixel_value1)
plt.show()
````

#### Step:2 画素値aから画素値bを255にする
- 前述の```np.where```を使用する
    ```np.where((before_pixel_value >= a) & (before_pixel_value <= b), 255, 0)```
- ```np.where((ピクセル変数 >= a) & (b <= ピクセル変数), x, y)``` と書くことで，画素値a以上b以下をxにし，そうで無い場合yにすると言う命令になる

````python
# 台形トーンカーブ
import numpy as np
import matplotlib.pyplot as plt

a = 25
b = 200
before_pixel_value = np.arange(256)
after_pixel_value2 = np.where((before_pixel_value >= a) & (before_pixel_value <= b), 255, 0)
after_pixel_value2 = after_pixel_value2.astype(np.uint8)

plt.plot(after_pixel_value2)
plt.show()
````

#### Step:3 画素値bから画素値255を255から0で直線変化
 - 画素値bから画素値255に対して，```-255 / (255 - b) * 画素値 + 255 * 255 / (255 - b)```で計算できる

````python
# 台形トーンカーブ
import numpy as np
import matplotlib.pyplot as plt

a = 25
b = 200

before_pixel_value = np.arange(256)
after_pixel_value3 = -255 / (255 - b) * before_pixel_value + 255 * 255 / (255 - b)
after_pixel_value3 = np.clip(after_pixel_value3, 0, 255)
after_pixel_value3 = after_pixel_value3.astype(np.uint8)

plt.plot(after_pixel_value3)
plt.show()
````

### Step4: それぞれの直線を組み合わせる
- 範囲```[0:a]```，```[a:b]```，```[b:256]```を組み合わせる
    - Step1: 0からa-1
    - Step2: aからb-1
    - Step3: bから255
- 組み合わせは```np.append```を使用する

````python
after_pixel_value = np.append(after_pixel_value1[0:a], after_pixel_value2[a:b])
after_pixel_value = np.append(after_pixel_value, after_pixel_value3[b:256])
print(after_pixel_value)
plt.plot(after_pixel_value)
plt.show()
````

````python
# 上下反転したい場合は255から画素値を引けば良い
after_pixel_value = 255 - after_pixel_value
plt.plot(after_pixel_value)
plt.show()
````

````python
# まとめると
# 台形トーンカーブ
import numpy as np
import matplotlib.pyplot as plt

a = 25
b = 200

before_pixel_value = np.arange(256)

# Step1
after_pixel_value1 = 255 / a * before_pixel_value
after_pixel_value1 = np.clip(after_pixel_value1, 0, 255)
after_pixel_value1 = after_pixel_value1.astype(np.uint8)

# Step2
after_pixel_value2 = np.where((before_pixel_value >= a) & (before_pixel_value <= b), 255, 0)
after_pixel_value2 = after_pixel_value2.astype(np.uint8)

# Step3
after_pixel_value3 = -255 / (255 - b) * before_pixel_value + 255 * 255 / (255 - b)
after_pixel_value3 = np.clip(after_pixel_value3, 0, 255)
after_pixel_value3 = after_pixel_value3.astype(np.uint8)

# Step4
after_pixel_value = np.append(after_pixel_value1[0:a], after_pixel_value2[a:b])
after_pixel_value = np.append(after_pixel_value, after_pixel_value3[b:256])
plt.plot(after_pixel_value)
plt.show()
````
