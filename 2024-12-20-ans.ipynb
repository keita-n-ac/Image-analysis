A. 以下の画像(wb-ex.png)に対して，各処理を行いなさい
- 3×3の十字のカーネルを使用すること

<img src="fig/wb-ex.jpg" width="50%">

1. 2回の膨張処理を行いなさい

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# カーネルの作成
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3, 3))

# 膨張処理（2回）
after_image = cv2.dilate(image, kernel, iterations=2)

plt.imshow(after_image)
plt.gray()
plt.show()
```

2. 2回の収縮処理を行いなさい

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# カーネルの作成
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3, 3))

# 収縮処理（2回）
after_image = cv2.erode(image, kernel, iterations=2)

plt.imshow(after_image)
plt.gray()
plt.show()
```

3. モルフォロジー勾配（2回）を行い，白黒反転（白を黒，黒を白に変換する）を行いなさい

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# カーネルの作成
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3, 3))

# モルフォロジー勾配
after_image = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel, iterations=2)
# 白黒反転
after_image2 = cv2.bitwise_not(after_image)


plt.imshow(after_image2)
plt.gray()
plt.show()
```

4. トリミングを行い『平』の文字だけを表示する画像を作成しなさい

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(image.shape)

# トリミング
after_image = image[70:230, 150:285]

plt.imshow(after_image)
plt.gray()
plt.show()
```

B. 以下の画像(wb-ex_black_noise.jpg)に対して，ノイズを除去しなさい
- 3×3の十字のカーネルを使用すること

<img src="fig/wb-ex_black_noise.jpg" width="50%">

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex_black_noise.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# カーネルの作成
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3, 3))

# クロージング処理
after_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel, iterations=3)

plt.imshow(after_image)
plt.gray()
plt.show()
```

C. 以下の画像(wb-ex_white_noise.jpg)に対して，ノイズを除去しなさい
- 3×3の十字のカーネルを使用すること

<img src="fig/wb-ex_white_noise.jpg" width="50%">

```python
import cv2
import matplotlib.pyplot as plt

# 画像の読み込み
image = cv2.imread('wb-ex_white_noise.jpg')

# BGR → GRAY
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# カーネルの作成
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3, 3))

# オープニング処理
after_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel, iterations=2)

plt.imshow(after_image)
plt.gray()
plt.show()
```
