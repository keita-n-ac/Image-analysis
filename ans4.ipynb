````python

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('q1.pgm') # ファイル名がq1.pgmの場合
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # BGR → グレー
plt.imshow(image)
plt.gray()
plt.show()
````
