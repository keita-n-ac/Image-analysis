### opencvの画像読み込み
- opencvライブラリである```cv2```をimportする
- ```変数名 = cv2.imread(画像ファイル)```を使用する
- ファイル名は文字列とする
- ファイルは同じフォルダに置いておく（講義の場合: デスクトップ）
```python
import cv2
image = cv2.imread(画像ファイル名)
```

````python
import cv2
image = cv2.imread('soya.jpeg')
````

### 変数に読み込みした画像の表示
- matplotライブラリを使用する
    - ```import matplotlib.pyplot as plt```でimportする
    - ```plt.imshow(変数名)```と```plt.show()```を使用する
```python
import matplotlib.pyplot as plt # ライブラリの導⼊
plt.imshow(変数名)
plt.show()
```
- まとめると以下のプログラムになる

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
plt.imshow(image)
plt.show()
````

### OpenCVで画像を読み込む際の注意
- OpenCVで画像を読み込む際，RGBの順番で画素値を読むのでは無く，**BGRの順番で画素値を読む**ため， **BGRからRGBとなるように，画素値を入れ替える必要がある**
- そこで，```cv2.cvtColor```を使用する
    - 色変換を行う
- 使い方，```変数名 = cv2.cvtColor(変数名, cv2.COLOR_BGR2RGB)```
    - ```cv2.COLOR_BGR2RGB```がBGRからRGBに変換する命令

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
````

- カラー画像からグレースケール画像に変更する際は，```cv2.cvtColor```を使用する
    - BGRからグレースケールに変換する```cv2.COLOR_BGR2GRAY```やRGBからグレースケールに変換する```cv2.COLOR_RGB2GRAY```を指定する 
- ただし，**matplotlibはカラー画像を標準で出力する**ため，```plt.gray()```を追加する

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
plt.imshow(image)
plt.gray()
plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # BGR → RGB
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) # RGB → グレー
plt.imshow(image)
plt.gray()
plt.show()
````

### opencvの画像の出力（保存）
- 変数に保存されている画素値の画像をファイル出力する
- ```cv2.imwrite(保存ファイル名, 変数名)```を使用する
- ファイル名は文字列とする
```python
import cv2
cv2.imread(保存ファイル名, 変数名)
```

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('soya.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
cv2.imwrite('result.jpeg', image)
plt.imshow(image)
plt.gray()
plt.show()
````

- 有名な画像フォーマット形式を扱えることができる
    - BMP形式（拡張子: bmp）
    - JPEG形式（拡張子: jpeg または jpg）
    - PNG形式（拡張子: png）
    - PGM形式（拡張子: pgm）
    - PPM形式（拡張子: ppm）
    - TIFF（拡張子: tiff または tif）


````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('test.pgm')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
cv2.imwrite('result.jpeg', image)
plt.imshow(image)
plt.gray()
plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('test.ppm')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR → RGB
cv2.imwrite('result.jpeg', image)
plt.imshow(image)
plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('nikka.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → GRAY
plt.imshow(image)
plt.show()
````

### PGMファイル
- 最後に1行（空白行）を付けること
- ファイル名を```sample.pgm```としてデスクトップに保存する
- 空白は半角スペースであけること
```
P2
5 5
255
128 0   255 0   128
0   128 255 128 0
255 255 255 255 255
0   128 255 128 0
128 0   255 0   128

```

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('sample.pgm')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR → グレー
cv2.imwrite('result.jpeg', image)
plt.imshow(image)
plt.gray()
plt.show()
````

### PPMファイル
- 最後に1行（空白行）を付けること
- ファイル名を```sample.ppm```としてデスクトップに保存する
- 空白は半角スペースであけること
```
P3
3 3
255
255 0 0 255 0 0 255 0 0
0 255 0 0 255 0 0 255 0
0 0 255 0 0 255 0 0 255

```

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('sample.ppm')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR → RGB
cv2.imwrite('result.jpeg', image)
plt.imshow(image)
plt.show()
````
